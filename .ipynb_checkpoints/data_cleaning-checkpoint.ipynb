```python
df['description_no_stopwords']=df['description'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
df['question_no_stopwords']=df['question'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in (stop)]))
```

```python
tokenized_lists_D=[]
for des in df['description_no_stopwords']:
#     word_tokens = word_tokenize(des)
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    new_words = tokenizer.tokenize(des)
    tokenized_lists_D.append(new_words)
    
```

```python
tokenized_lists_D
```

```python
df["d_tokenized_no_stopwords"]=tokenized_lists_D
```

```python
df["d_tokenized_no_stopwords"]
```

```python
tokenized_lists_Q=[]
for des in df['question_no_stopwords']:
#     word_tokens = word_tokenize(des)
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    new_words = tokenizer.tokenize(des)
    tokenized_lists_Q.append(new_words)
    
```

```python

def krovetz_stem(listy):
    ks = krovetz.PyKrovetzStemmer()
    stemmed_list=[]
    for l in listy:
        new_list=[]
        for word in l:
#             print(word)
#             print(ks.stem(word))
            new_list.append(ks.stem(word))
        stemmed_list.append(new_list)
    return stemmed_list
            
        
        
```

```python
tokenize_stemmed_stopword_list_D = krovetz_stem(tokenized_lists_D)
tokenize_stemmed_stopword_list_Q = krovetz_stem(tokenized_lists_Q)
```

```python
df["q_tokenized_stemmed_no_stopwords"]=tokenize_stemmed_stopword_list_Q
df["d_tokenized_stemmed_no_stopwords"]=tokenize_stemmed_stopword_list_D
```

```python
df["q_tokenized_stemmed_no_stopwords"]
```

```python
type(df["d_tokenized_stemmed_no_stopwords"][45])
```

```python
#applying tfidf using values of df["d_tokenized_stemmed_no_stopwords"] as columns and df["q_tokenized_stemmed_no_stopwords"] as columns
#create a set of words from the descrition that will become our columns 
columns_tfidf = set()
for row in tokenize_stemmed_stopword_list_D:
    for word in row:
        columns_tfidf.add(word)
```

```python
columns_tfidf
```

```python
tokenize_stemmed_stopword_list_Q 
```

```python
#calculate document frequency for each word 
DF = {}
for word in columns_tfidf:
    for i in range(len(tokenize_stemmed_stopword_list_Q)):
           if(word in tokenize_stemmed_stopword_list_Q[i]):
                try:
                    DF[word].add(i)
                except:
                    DF[word] = {i}
```

```python
DF
```

```python
iDF={}
for i in DF:
    DF[i] =  len(DF[i])
    iDF[i]= np.log(len(tokenize_stemmed_stopword_list_Q)/(DF[i]+1))
```

```python
DF #how many times the word has occured in the question
```

```python
ks = krovetz.PyKrovetzStemmer()
ks.stem('numbers')
```

```python
iDF
```

```python
#cal the term freq for each question
term_freq_dict = dict() 

for index in range(len(tokenize_stemmed_stopword_list_Q)):
    for word in columns_tfidf:
        if word in tokenize_stemmed_stopword_list_Q[index]:
            term_freq_dict.setdefault(index, []).append((word,tokenize_stemmed_stopword_list_Q[index].count(word)))
               
```

```python
term_freq_dict
```

```python
tf_idf=np.array([len(tokenize_stemmed_stopword_list_Q),len(columns_tfidf)])
```

```python
tf_idf=dict() 
for key,value in term_freq_dict.items():
    for val in value:
        tf_idf.setdefault(key, []).append((val[0],iDF[val[0]]*val[1]))
        
```
